```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import math

def step(x):
    y=x>0
    return np.array(x>0,dtype=np.int)

e=math.e

def sigmoid(x):
    sig=1/(1+e**-x)
    return sig

x=np.arange(-5,5,0.01)

y=sigmoid(x)

plt.plot(x,y)

y2=step(x)

plt.plot(x,y2)

plt.grid()
plt.xlabel("x", fontsize=14)
plt.ylabel("sigmoid(x)", fontsize=14)
plt.show()
```
