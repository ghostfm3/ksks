```python
import matplotlib.pyplot as plt
%matplotlib inline
import gensim
import nltk

nltk.download("reuters")
nltk.download("stopwords")
nltk.download("punkt")
```

```python
reuters = nltk.corpus.reuters
paras = reuters.paras()
```

```python
paras[0][0][:20]
```

```python
stop_words = set(nltk.corpus.stopwords.words("english"))
```

```python
sorted(stop_words)[:10]
```

```python
import itertools
reuters_texts = [itertools.chain(*p) for p in paras]
reuters_texts_filtered = [[w.lower() for w in p if w not in stop_words and len(w) >= 2] for p in reuters_texts]
```

```python
dictionary = gensim.corpora.Dictionary(itertools.chain(reuters_texts_filtered))
reuters_corpus = [dictionary.doc2bow(text) for text in reuters_texts_filtered]
```

```python
import nltk
from nltk import sent_tokenize, word_tokenize
import urllib
import urllib.request
```

```python
urllib.request.urlretrieve('http://www.gutenberg.org/files/74/74-0.txt',"tomsawyer.txt")
with open("tomsawyer.txt",'r',encoding='utf-8-sig',errors='ignore') as f:
    print(f.read())
```

```python
text_law = open("tomsawyer.txt","r",encoding='UTF-8',errors='ignore').read()
text_sent = sent_tokenize(text_law)

text_word = []

for t in text_sent:
    text_word.append(word_tokenize(t.lower()))
```

```python
print(text_word[0:2])
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(text_word,size=300,window=5,min_count=5,workers=4)
```

```python
model.train(text_word,total_examples=len(text_word),epochs=100)
```

```python
model.wv["cost"]
```

```python
model.wv.most_similar("under")
```

```python
from gensim.test.utils import datapath
```

```python
analogy_scores = model.wv.evaluate_word_analogies(datapath('questions-words.txt'))
```
